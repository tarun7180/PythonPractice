```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
carsData = pd.read_csv('Toyota.csv', index_col=0, na_values=["??","???"])
```

```python
carsData.size
```

```python
carsData.dropna(axis = 0, inplace = True)
```

```python
carsData.size
```

```python
#scatter plot with default arguments
sns.set(style="darkgrid")
sns.regplot(x=carsData['Age'], y=carsData['Price'])
```

```python
#scatter plot without the regression fit line
sns.regplot(x=carsData['Age'], y=carsData['Price'], fit_reg=False)
```

```python
#scatter plot by customizing the appearance of markers
sns.regplot(x=carsData['Age'], y=carsData['Price'], marker="*", fit_reg=False)
```

```python
#scatter plot of price vs age by FuelType
#Using hue parameter, inclduing another variable to sho wthe fuel type categories with different colors
sns.lmplot(x='Age',
          y = 'Price',
          data = carsData,
          fit_reg = False,
          hue = 'FuelType',
          legend = True,
          palette = "Set1")
```

```python
#Histogram with default kernel density estimate
sns.displot(carsData['Age'], kde=True)
```

```python
#Histogram without kernel density estimate
sns.displot(carsData['Age'],kde=False)
```

```python
#Histogram with fixed number of bins
sns.displot(carsData['Age'], kde = False, bins = 5)
```

```python
#Bar plot - frequency distribution of the fuel type of the cars
sns.countplot(x="FuelType", data=carsData)
```

```python
# Grouped bar plot of FuelType and Automatic
sns.countplot(x="FuelType", data=carsData,hue="Automatic")
```

```python
#box and whisker plot of Price
sns.boxplot(y=carsData["Price"])
```

```python
#Box and whisker plot for numerical vs categorical variable. Price os the car for various fuel types
sns.boxplot(x = carsData['FuelType'], y = carsData['Price'])
```

```python
#Grouped box and whisker plot of Price vs FuelType and Automatic
sns.boxplot(x = "FuelType",
           y = carsData["Price"],
           hue = 'Automatic',
           data = carsData)
```

```python
#box-whisker plot and histogram on the same window
f,(ax_box, ax_hist)=plt.subplots(2, gridspec_kw={"height_ratios":(.15, .85)})
sns.boxplot(carsData["Price"], ax=ax_box)
sns.histplot(carsData["Price"], ax=ax_hist, kde=False)
```

```python
#Pairwise plots - plot pairwise relationships in a dataset
#Create scatterplots for joint relationships and histograms for univariate distributions
sns.pairplot(carsData, kind="scatter", hue="FuelType")
plt.show()
```
